```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from google.colab import drive
drive.mount("/content/drive")
```

```python
file = open("/content/drive/MyDrive/Orbit-Data-25544-GP.csv")
df = pd.read_csv(file)
```

```python
# --- 데이터 준비 ---
df['Date/Time (UTC)'] = pd.to_datetime(df['Date/Time (UTC)'])

# 데이터를 시간순으로 정렬
df = df.sort_values(by='Date/Time (UTC)')

# --- 그래프 그리기 ---
plt.figure(figsize=(14, 7))  # 그래프 크기를 보기 좋게 키움

# 꺾은선 그래프 생성
plt.plot(df['Date/Time (UTC)'], df['SMA'], linestyle='-', linewidth=1)

# --- 그래프 디테일 설정 ---
plt.title('ISS Orbit Decay (SMA vs. Time)', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Semi-Major Axis (SMA) [km]', fontsize=12)
plt.grid(True)

# X축 날짜 형식이 겹치지 않게 '년-월' 형식으로 자동 조절
ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gcf().autofmt_xdate()

plt.tight_layout()  # 그래프 여백 자동 조절
plt.show();          # 최종 그래프 출력
```

```python
# --- 4. 데이터 준비 (이전과 동일) ---
df['Date/Time (UTC)'] = pd.to_datetime(df['Date/Time (UTC)'])
df = df.sort_values(by='Date/Time (UTC)')

# --- 5. 이심률(Eccentricity) 그래프 그리기 ---
plt.figure(figsize=(14, 7))
plt.plot(df['Date/Time (UTC)'], df['Eccentricity'], linestyle='-', linewidth=1, color='orange')

# --- 6. 그래프 디테일 설정 ---
plt.title('ISS Orbit Eccentricity vs. Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Eccentricity', fontsize=12)
plt.grid(True)

ax = plt.gca()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gcf().autofmt_xdate()

plt.tight_layout()
plt.show();
```
